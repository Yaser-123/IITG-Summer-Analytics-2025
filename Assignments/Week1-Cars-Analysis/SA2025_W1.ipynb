### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
from colorama import Fore, Style, init

init(autoreset=True)

median_mpg = df['mpg'].median()
above_median = df[df['mpg'] > median_mpg]
best_ratio = above_median['horsepower'] / above_median['weight']
top3_indices = best_ratio.nlargest(3).index

print(f"{Fore.GREEN} Top 3 Cars by Horsepower/Weight Ratio (mpg > median):{Style.RESET_ALL}\n")

for i, name in enumerate(top3_indices, 1):
    ratio = best_ratio.loc[name]
    if isinstance(ratio, pd.Series):
        ratio = ratio.iloc[0]
    print(f"{Fore.YELLOW}{i}. {Style.BRIGHT}{name}{Style.NORMAL} {Fore.CYAN} Ratio: {ratio:.6f}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set emoji-supporting font first
plt.rcParams['font.family'] = 'Segoe UI Emoji'

# Set Seaborn style
sns.set_style('whitegrid')

# Process data
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
pivoted = avg_mpg.pivot(index='model_year', columns='origin', values='mpg')

# Plot
plt.figure(figsize=(12, 7))
for origin in pivoted.columns:
    plt.plot(pivoted.index, pivoted[origin],
             marker='o', linewidth=2.5, label=origin)

# Labels and legend
plt.title(
    'Average MPG Over Years by Origin ',
    fontdict={
        'fontsize': 18,
        'fontweight': 'bold',
        'family': 'Segoe UI Emoji'
    }
)
plt.xlabel('Model Year', fontsize=14)
plt.ylabel('Average MPG', fontsize=14)
plt.legend(title='Origin', title_fontsize=13, fontsize=12)

plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.set_style('whitegrid')

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    alpha=0.7,
    sizes=(20, 200)  # optional: adjust size range
)

plt.title(
    'Horsepower vs Weight by Origin (Size = MPG)',
    fontdict={'fontsize': 16, 'fontweight': 'bold', 'family': 'Segoe UI Emoji'}
)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Origin & MPG')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
name_counts = df.reset_index().groupby('name').size()
multi_year = name_counts[name_counts > 1].index
multi_year = name_counts[name_counts > 1].index
mpg_std = df.groupby('name')['mpg'].std()
consistent_models = mpg_std[(mpg_std < 1.0) & (mpg_std.index.isin(multi_year))]

result = df.groupby('name').agg(
    appearances=('mpg', 'count'),
    avg_mpg=('mpg', 'mean'),
    mpg_std=('mpg', 'std')
).loc[consistent_models.index]

result = result[result['mpg_std'] < 1.0]
result = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])
result[['appearances', 'avg_mpg']]
```
