```python
# Import additional libraries for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# FAST parameter grid - optimized for speed while maintaining performance
param_grid = {
    'n_estimators': [500, 1000],        # Reduced from 3 to 2 options
    'max_depth': [None, 30],            # Reduced from 3 to 2 options  
    'min_samples_split': [2],           # Fixed at optimal value
    'min_samples_leaf': [1],            # Fixed at optimal value
    'max_features': ['sqrt'],           # Fixed at optimal value
    'bootstrap': [True],                # Fixed at optimal value
    'random_state': [42]
}

print("üîç Starting FAST hyperparameter tuning...")
print(f"Parameter combinations to test: {len(param_grid['n_estimators']) * len(param_grid['max_depth'])}")

# Create Random Forest classifier
rf_tuned = RandomForestClassifier()

# Perform grid search with cross-validation - reduced CV folds for speed
grid_search = GridSearchCV(
    estimator=rf_tuned,
    param_grid=param_grid,
    cv=3,                    # Reduced from 5 to 3 for speed
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search
grid_search.fit(X_train, y_train)

print(f"\n‚úÖ Best parameters found:")
for param, value in grid_search.best_params_.items():
    print(f"   {param}: {value}")

print(f"\nüéØ Best cross-validation accuracy: {grid_search.best_score_:.4f}")

# Get the best model
best_model = grid_search.best_estimator_
```

## üìä Step 6: Strategic Train-Test Split

### üé≤ **Data Splitting Strategy:**
Create robust train-test split with:
- **80% training data** for model learning
- **20% test data** for unbiased evaluation
- **Stratified sampling** to maintain class balance
- **Random state** for reproducible results

```python
from sklearn.ensemble import RandomForestClassifier
import time

print("üå≤ Initializing optimized Random Forest classifier...")

# Create Random Forest with optimized parameters for maximum accuracy
model = RandomForestClassifier(
    n_estimators=200,          # Increased trees for better performance
    max_depth=20,              # Controlled depth to prevent overfitting
    min_samples_split=2,       # Allow splitting with minimum samples
    min_samples_leaf=1,        # Allow small leaves for detailed learning
    max_features='sqrt',       # Optimal feature randomness
    bootstrap=True,            # Use bootstrap sampling
    random_state=42,           # Reproducible results
    n_jobs=-1,                 # Use all CPU cores for faster training
    verbose=0                  # Clean output
)

print(f"‚úÖ Model configured with optimized parameters:")
print(f"   ‚Ä¢ Estimators: {model.n_estimators} trees")
print(f"   ‚Ä¢ Max depth: {model.max_depth}")
print(f"   ‚Ä¢ Feature selection: {model.max_features}")
print(f"   ‚Ä¢ Bootstrap: {model.bootstrap}")

print(f"\nüöÄ Training Random Forest on {X_train.shape[0]} samples...")
start_time = time.time()

# Train the model
model.fit(X_train, y_train)

training_time = time.time() - start_time

print(f"‚úÖ Model training completed successfully!")
print(f"‚è±Ô∏è Training time: {training_time:.2f} seconds")
print(f"üìä Model trained on {X_train.shape[0]} samples with {X_train.shape[1]} features")
print(f"üéØ Ready for evaluation and predictions!")
```

## üå≤ Step 7: Optimized Random Forest Model Training

### üéØ **Advanced Random Forest Configuration:**
Implement an optimized Random Forest with parameters tuned for **maximum accuracy**:

#### üîß **Key Optimizations:**
- **Increased Estimators**: 200 trees (vs. default 100) for better ensemble strength
- **Controlled Depth**: max_depth=20 to prevent overfitting while capturing complexity
- **Bootstrap Sampling**: Default True for variance reduction
- **Feature Randomness**: sqrt(n_features) for optimal feature selection

```python
# üìä COMPREHENSIVE MODEL EVALUATION
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("üìä Evaluating model performance on test set...")

# Calculate accuracy on test set
test_accuracy = model.score(X_test, y_test)

# Generate predictions for detailed metrics
y_pred = model.predict(X_test)

# Calculate comprehensive metrics
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Training accuracy for comparison
train_accuracy = model.score(X_train, y_train)

print(f"üéØ MODEL PERFORMANCE RESULTS:")
print(f"{'='*50}")
print(f"üìà Test Accuracy:      {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"üìà Training Accuracy:  {train_accuracy:.4f} ({train_accuracy*100:.2f}%)")
print(f"üìä Precision (weighted): {precision:.4f}")
print(f"üìä Recall (weighted):    {recall:.4f}")
print(f"üìä F1-Score (weighted):  {f1:.4f}")

# Performance analysis
print(f"\nüîç PERFORMANCE ANALYSIS:")
if test_accuracy >= 1.0:
    print(f"üèÜ PERFECT SCORE! Achieved 100% accuracy!")
elif test_accuracy >= 0.99:
    print(f"ü•á EXCELLENT! Near-perfect performance!")
elif test_accuracy >= 0.95:
    print(f"ü•à VERY GOOD! High accuracy achieved!")
else:
    print(f"ü•â GOOD! Room for improvement with hyperparameter tuning.")

# Overfitting check
overfitting = train_accuracy - test_accuracy
print(f"\nüîç Overfitting Analysis:")
print(f"   ‚Ä¢ Accuracy gap: {overfitting:.4f}")
if overfitting < 0.01:
    print(f"   ‚úÖ Excellent generalization!")
elif overfitting < 0.05:
    print(f"   ‚úÖ Good generalization!")
else:
    print(f"   ‚ö†Ô∏è Potential overfitting detected!")

print(f"\nüìä Model successfully evaluated on {len(y_test)} test samples!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

print("üìä Generating confusion matrix for detailed analysis...")

# Generate confusion matrix
cm = confusion_matrix(y_test, y_predicted)

print(f"‚úÖ Confusion matrix generated!")
print(f"üìä Matrix shape: {cm.shape}")

print(f"\nüîç CONFUSION MATRIX:")
print(cm)

print(f"\nüìà PER-CLASS PERFORMANCE ANALYSIS:")
print(f"{'Digit':<6} {'Samples':<8} {'Correct':<8} {'Accuracy':<10}")
print(f"{'='*35}")

for i in range(10):
    class_total = cm[i].sum()
    class_correct = cm[i, i]
    class_accuracy = class_correct / class_total if class_total > 0 else 0
    print(f"{i:<6} {class_total:<8} {class_correct:<8} {class_accuracy:<10.4f}")

# Overall matrix analysis
diagonal_sum = np.trace(cm)
total_predictions = cm.sum()
matrix_accuracy = diagonal_sum / total_predictions

print(f"\nüìä MATRIX SUMMARY:")
print(f"   ‚Ä¢ Total predictions: {total_predictions}")
print(f"   ‚Ä¢ Correct predictions: {diagonal_sum}")
print(f"   ‚Ä¢ Matrix accuracy: {matrix_accuracy:.4f}")
print(f"   ‚Ä¢ Perfect diagonal: {np.all(cm == np.diag(np.diag(cm)))}")

# Check for misclassifications
off_diagonal_sum = cm.sum() - np.trace(cm)
print(f"\nüîç MISCLASSIFICATION ANALYSIS:")
print(f"   ‚Ä¢ Total misclassifications: {off_diagonal_sum}")

if off_diagonal_sum == 0:
    print(f"   üéâ PERFECT CLASSIFICATION! No errors detected!")
else:
    print(f"   ‚ùå Found {off_diagonal_sum} misclassifications")
    # Show most common confusions
    cm_off_diag = cm.copy()
    np.fill_diagonal(cm_off_diag, 0)
    if cm_off_diag.max() > 0:
        max_confusion = np.unravel_index(cm_off_diag.argmax(), cm_off_diag.shape)
        print(f"   ‚Ä¢ Most confused: Digit {max_confusion[0]} ‚Üí {max_confusion[1]} ({cm_off_diag[max_confusion]} times)")

print(f"\n‚úÖ Confusion matrix analysis completed!")
```
